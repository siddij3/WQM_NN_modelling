```python
import pandas as pd
import os


local_download_path = os.path.expanduser('.\\')
filenames=[]
for filename in os.listdir(local_download_path):
    if filename.endswith('csv') :#and 'Entries' in filename:
      filenames.append(filename)
      
print(filenames)

df = pd.read_csv(filenames[0], header=None)


edge_ls = [1 if j < 10 else 0 for i in range(20) for j in range(50)]
start_ls = [1 if 9 < j < 30 else 0 for i in range(20) for j in range(50)]
end_ls = [1 if j > 29 else 0 for i in range(20) for j in range(50)]
pH_ls = [7 if not (j % 2) else 10 if j == 1 else 3 if j ==
         3 else None for i in range(5) for j in range(4) for k in range(50)]
```

```python


days_elapsed = df.values.tolist()[0][1:]
sensors_number = df.values.tolist()[1][1:]
df2 = df[3:]

times=[]
voltage=[]
currents=[]


t = 1
for i,j in df2.iterrows():
  
  curr = list(map(float,j.values))
  
  times.append(t)
  if (curr[0] % 50 == 0):
    t = 0
  
  currents.append(curr[1:])
  t+=1
  
  for i in curr[1:]:
    voltage.append(i)

length=len(currents[19])

durations = []
edges = []
starts = []
ends = []
pH = []


n=-1

for i in range(len(voltage)):
  voltage[i]
  if i%length==0:
    n+=1

  durations.append(times[n])

  edges.append(edge_ls[n])
  starts.append(start_ls[n])
  ends.append(end_ls[n])
  pH.append(pH_ls[n])

print(durations)

days = []
num_sensors = []

m=0
for i in range(len(voltage)):
  if(m==len(days_elapsed)): 
    m=0
  days.append(days_elapsed[m])
  num_sensors.append(sensors_number[m])

  m+=1

dict3 = {'Time':durations, 'Voltage':voltage, 'Days Elapsed':days ,'Sensor Number':num_sensors, 'Rising Edge':edges, 'Starting Edge':starts, 'End Measurement':ends , 'pH':pH}
df_final = pd.DataFrame(dict3)

df_final

df_final.to_csv('organized_data_2.csv',index=False)

```
